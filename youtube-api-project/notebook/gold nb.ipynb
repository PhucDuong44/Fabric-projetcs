#### **tạo bảng dim**

```python
df_silver = spark.read.table('silver_youtube_final')
# Tạo Dim Keyword
df_silver.select("keyword").distinct().write.mode("overwrite").saveAsTable("dim_keyword")

# Tạo Dim Date
from pyspark.sql.functions import to_date
df_silver.select(to_date("fetched_at").alias("date")).distinct().write.mode("overwrite").saveAsTable("dim_date")
```
